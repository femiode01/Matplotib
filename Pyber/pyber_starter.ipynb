# Observable Trends

1) From the data provided, it shows that fares are much more higher in urban cities with a percentage of 62.7% compared to suburban cities with a percentage of 30.5% and Rural Areas with 6.8%.

2) The total rides per city analysis shows that people take rides almost 3 times more in brban cities compared to suburban and rural areas.

3) The data also shows that majority of the drivers are in the Urban cities compared to 16.5% in the suburban areas and 2.6% in rural areas.

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
df_city = pd.read_csv(city_data_to_load)
df_ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
starter_py = pd.merge(df_ride, df_city, how="inner", on="city")

# Display the data table for preview
starter_py.head()
```

```python

urbanc= starter_py[starter_py["type"]=="Urban"]
suburbanc= starter_py[starter_py["type"]=="Suburban"]
ruralc= starter_py[starter_py["type"]=="Rural"]

average_fare_u = urbanc.groupby(['city']).mean()['fare']
rides_per_city_u= urbanc.groupby(['city']).count()['ride_id']
tot_drivers_u = urbanc.groupby(['city']).sum()['driver_count']


average_fare_s = suburbanc.groupby(['city']).mean()['fare']
rides_per_city_s= suburbanc.groupby(['city']).count()['ride_id']
tot_drivers_s = suburbanc.groupby(['city']).sum()['driver_count']



average_fare_r = ruralc.groupby(['city']).mean()['fare']
rides_per_city_r= ruralc.groupby(['city']).count()['ride_id']
tot_drivers_r = ruralc.groupby(['city']).sum()['driver_count']
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(rides_per_city_u, average_fare_u, s=15*rides_per_city_u, marker="o", linewidth=1.0, facecolors="coral", edgecolors="black", label= "Urban")
plt.scatter(rides_per_city_s, average_fare_s, s=12*rides_per_city_s, marker="o",linewidth=1.0, facecolors="lightblue", edgecolors="black", label= "Suburban")
plt.scatter(rides_per_city_r, average_fare_r, s=10*rides_per_city_r, marker="o",linewidth=1.0, facecolors="yellow", edgecolors="black", label= "Rural")


# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare (Dollars)")

# Create a legend
legend = plt.legend(fontsize="small", title="City Types")

legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

plt.xlim((0,41))
plt.grid(True)



# Incorporate a text label regarding circle size

t = (" Note: Circle size correlates with driver count by city")
plt.text(58, 35, t, ha='left', horizontalalignment='center', verticalalignment='center')

# Save Figure
plt.savefig("../Images/Pyberrideshare.png")
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
percent_city_type_u = starter_py.groupby(['type']).sum()['fare']

# Build Pie Chart
city = ["Rural", "Surburban", "Urban",]
colors = ["Yellow", "Lightblue", "coral"]
plt.title("% of Total Fares by City Type")
plt.pie(percent_city_type_u, labels=city, explode=(0,0,0.1), colors=colors, autopct="%1.1f%%", shadow=False, startangle=140)

# Save Figure
plt.savefig("../Images/Percenttotfaresbycity.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
percent_rides_type_u = starter_py.groupby(["type"]).count()['ride_id']

# Build Pie Chart

city = ["Rural", "Surburban", "Urban",]
colors = ["Yellow", "Lightblue", "coral"]
plt.title("% of Total Rides by City Type")
plt.pie(percent_rides_type_u, labels=city, explode=(0,0,0.1), colors=colors, autopct="%1.1f%%", shadow=False, startangle=140)
# Save Figure
plt.savefig("../Images/Percenttotridesbycity.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

dupli_dri = starter_py.drop_duplicates(['city'], keep='first')
percent_drivers_type_u =dupli_dri.groupby(["type"]).sum()['driver_count']
percent_drivers_type_u

# Build Pie Charts
city = ["Rural", "Surburban", "Urban",]
colors = ["Yellow", "Lightblue", "coral"]
plt.title("% of Total Drivers by City Type")
plt.pie(percent_drivers_type_u, labels=city, explode=(0,0,0.1), colors=colors, autopct="%1.1f%%", shadow=False, startangle=140)
# Save Figure
plt.savefig("../Images/Percentdriversbycity.png")
# Show Figure
plt.show()
```
